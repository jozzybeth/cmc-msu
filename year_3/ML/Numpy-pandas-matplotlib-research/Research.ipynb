```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge

class EchoStateNetwork:
    def __init__(self, n_inputs, n_reservoir, n_outputs, spectral_radius=0.9, sparsity=0.2):
        self.n_inputs = n_inputs
        self.n_reservoir = n_reservoir
        self.n_outputs = n_outputs

        self.W_in = np.random.uniform(-1, 1, (n_reservoir, n_inputs))
        self.W = np.random.uniform(-1, 1, (n_reservoir, n_reservoir)) #
        mask = np.random.rand(*self.W.shape) > sparsity # 
        self.W[mask] = 0

        eigvals = np.linalg.eigvals(self.W)
        self.W *= spectral_radius / np.max(np.abs(eigvals)) #   

        self.state = np.zeros((n_reservoir, 1)) # 

    def _update_state(self, input_vector): #   
        self.state = np.tanh(np.dot(self.W, self.state) + np.dot(self.W_in, input_vector))
      

    def fit(self, inputs, outputs, reg=1e-4): #   
        states = []
        for input_vector in inputs:
            self._update_state(input_vector)
            states.append(self.state.flatten())

        states = np.array(states)
        states = np.hstack((states, np.ones((states.shape[0], 1))))  #  

        ridge = Ridge(alpha=reg) #        W_out
        ridge.fit(states, outputs)
        self.W_out = ridge.coef_

    def predict(self, input_sequence): #    W_out
        predictions = []
        for input_vector in input_sequence:
            self._update_state(input_vector)
            state_with_bias = np.hstack((self.state.flatten(), [1]))  #  
            prediction = np.dot(self.W_out, state_with_bias)
            predictions.append(prediction)
        return np.array(predictions)

n_samples = 2000
n_predict = 700
n_reservoir = 1000
spectral_radius = 0.8

x = np.linspace(0, 20 * np.pi, n_samples)
y = np.sin(x) + np.sin(1.65*x) 

esn = EchoStateNetwork(n_inputs=1, n_reservoir=n_reservoir, n_outputs=1, spectral_radius=spectral_radius)
esn.fit(y[:-n_predict].reshape(-1, 1), y[1:-n_predict+1].reshape(-1, 1))

predictions = []
input_value = y[-n_predict].reshape(1, -1)

for _ in range(n_predict):
    prediction = esn.predict(input_value)
    predictions.append(prediction[0])
    input_value = prediction.reshape(1, -1)

x_true = x[:-n_predict]
y_true = y[:-n_predict]
plt.figure(figsize=(12, 6))
plt.plot(x_true, y_true, label='', color='blue')

#       x,     
x_predict = x[-n_predict:]

plt.plot(x_predict, predictions, label=' ESN', color='red')

plt.legend()
plt.title('     ESN')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

class ReservoirNetwork:
    def __init__(self, n_inputs, n_reservoir, n_outputs, spectral_radius=0.95, sparsity=0.1, seed=None):
        np.random.seed(seed)
        
        #  
        self.n_inputs = n_inputs
        self.n_reservoir = n_reservoir
        self.n_outputs = n_outputs
        
        #      
        self.Win = np.random.uniform(-1, 1, (n_reservoir, n_inputs))
        self.W = np.random.rand(n_reservoir, n_reservoir) - 0.5
        self.W *= np.random.rand(*self.W.shape) < sparsity  # 
        
        #    
        radius = max(abs(np.linalg.eigvals(self.W)))
        self.W *= spectral_radius / radius
        
        #   
        self.Wout = np.zeros((n_outputs, n_reservoir))
        
        #  
        self.state = np.zeros((n_reservoir, 1))
    
    def _update_state(self, u):
        #   
        pre_activation = np.dot(self.Win, u) + np.dot(self.W, self.state)
        self.state = np.tanh(pre_activation)
    
    def train(self, inputs, outputs, reg=1e-8):
        #     
        states = []
        
        #    
        for u in inputs:
            u = u.reshape(-1, 1)
            self._update_state(u)
            states.append(self.state.copy())
        
        #   
        states = np.hstack(states)
        
        #      
        self.Wout = np.dot(outputs.T, states.T).dot(np.linalg.inv(states.dot(states.T) + reg * np.eye(self.n_reservoir)))
    
    def predict(self, inputs):
        # 
        predictions = []
        
        for u in inputs:
            u = u.reshape(-1, 1)
            self._update_state(u)
            y = np.dot(self.Wout, self.state)
            predictions.append(y.ravel())
        
        return np.array(predictions)

#  
n_inputs = 1
n_reservoir = 100
n_outputs = 1

#  
reservoir = ReservoirNetwork(n_inputs, n_reservoir, n_outputs)

#   
t = np.linspace(0, 50, 500)
data = np.sin(t)

#    
train_length = 400
test_length = 100

#      
train_inputs = data[:train_length].reshape(-1, 1)
train_outputs = data[1:train_length + 1].reshape(-1, 1)

# 
reservoir.train(train_inputs, train_outputs)

# 
test_inputs = data[train_length:train_length + test_length].reshape(-1, 1)
predictions = reservoir.predict(test_inputs)

#  
plt.figure(figsize=(10, 5))
plt.plot(t[train_length:train_length + test_length], data[train_length:train_length + test_length], label="  ")
plt.plot(t[train_length + 1:train_length + test_length + 1], predictions, label=" ESN")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.legend()
plt.show()
```
