```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge

class EchoStateNetwork:
    def __init__(self, n_inputs, n_reservoir, n_outputs, spectral_radius=0.9, sparsity=0.2):
        self.n_inputs = n_inputs
        self.n_reservoir = n_reservoir
        self.n_outputs = n_outputs

        self.W_in = np.random.uniform(-1, 1, (n_reservoir, n_inputs))
        self.W = np.random.uniform(-1, 1, (n_reservoir, n_reservoir)) #резервуар
        mask = np.random.rand(*self.W.shape) > sparsity #разреженность матрицы
        self.W[mask] = 0

        eigvals = np.linalg.eigvals(self.W)
        self.W *= spectral_radius / np.max(np.abs(eigvals)) # нормировка матрицы резервуара

        self.state = np.zeros((n_reservoir, 1)) #состояние резервуара

    def _update_state(self, input_vector): #обновление внутреннего состояния резервуара
        self.state = np.tanh(np.dot(self.W, self.state) + np.dot(self.W_in, input_vector))
      

    def fit(self, inputs, outputs, reg=1e-4): #обучение через состояние сети
        states = []
        for input_vector in inputs:
            self._update_state(input_vector)
            states.append(self.state.flatten())

        states = np.array(states)
        states = np.hstack((states, np.ones((states.shape[0], 1))))  # добавляем смещение

        ridge = Ridge(alpha=reg) # Используем линейную регрессию с регуляризацией для вычисления W_out
        ridge.fit(states, outputs)
        self.W_out = ridge.coef_

    def predict(self, input_sequence): #прогноз через сформировавшуюся матрицу W_out
        predictions = []
        for input_vector in input_sequence:
            self._update_state(input_vector)
            state_with_bias = np.hstack((self.state.flatten(), [1]))  # добавляем смещение
            prediction = np.dot(self.W_out, state_with_bias)
            predictions.append(prediction)
        return np.array(predictions)

n_samples = 2000
n_predict = 700
n_reservoir = 1000
spectral_radius = 0.8

x = np.linspace(0, 20 * np.pi, n_samples)
y = np.sin(x) + np.sin(1.65*x) 

esn = EchoStateNetwork(n_inputs=1, n_reservoir=n_reservoir, n_outputs=1, spectral_radius=spectral_radius)
esn.fit(y[:-n_predict].reshape(-1, 1), y[1:-n_predict+1].reshape(-1, 1))

predictions = []
input_value = y[-n_predict].reshape(1, -1)

for _ in range(n_predict):
    prediction = esn.predict(input_value)
    predictions.append(prediction[0])
    input_value = prediction.reshape(1, -1)

x_true = x[:-n_predict]
y_true = y[:-n_predict]
plt.figure(figsize=(12, 6))
plt.plot(x_true, y_true, label='Синусоида', color='blue')

# Для предсказаний используем тот же диапазон x, продолжая от последней точки синусоиды
x_predict = x[-n_predict:]

plt.plot(x_predict, predictions, label='Предсказания ESN', color='red')

plt.legend()
plt.title('Генерация синусоидального сигнала с помощью ESN')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

class ReservoirNetwork:
    def __init__(self, n_inputs, n_reservoir, n_outputs, spectral_radius=0.95, sparsity=0.1, seed=None):
        np.random.seed(seed)
        
        # Параметры сети
        self.n_inputs = n_inputs
        self.n_reservoir = n_reservoir
        self.n_outputs = n_outputs
        
        # Инициализация весов входного и резервуарного слоя
        self.Win = np.random.uniform(-1, 1, (n_reservoir, n_inputs))
        self.W = np.random.rand(n_reservoir, n_reservoir) - 0.5
        self.W *= np.random.rand(*self.W.shape) < sparsity  # Спарсность
        
        # Нормировка на спектральный радиус
        radius = max(abs(np.linalg.eigvals(self.W)))
        self.W *= spectral_radius / radius
        
        # Инициализация выходного слоя
        self.Wout = np.zeros((n_outputs, n_reservoir))
        
        # Состояние сети
        self.state = np.zeros((n_reservoir, 1))
    
    def _update_state(self, u):
        # Обновление состояния резервуара
        pre_activation = np.dot(self.Win, u) + np.dot(self.W, self.state)
        self.state = np.tanh(pre_activation)
    
    def train(self, inputs, outputs, reg=1e-8):
        # Тренировка выходного слоя с регуляризацией
        states = []
        
        # Прогонка данных через резервуар
        for u in inputs:
            u = u.reshape(-1, 1)
            self._update_state(u)
            states.append(self.state.copy())
        
        # Сборка матрицы состояний
        states = np.hstack(states)
        
        # Вычисление выходных весов через линейную регрессию
        self.Wout = np.dot(outputs.T, states.T).dot(np.linalg.inv(states.dot(states.T) + reg * np.eye(self.n_reservoir)))
    
    def predict(self, inputs):
        # Прогнозирование
        predictions = []
        
        for u in inputs:
            u = u.reshape(-1, 1)
            self._update_state(u)
            y = np.dot(self.Wout, self.state)
            predictions.append(y.ravel())
        
        return np.array(predictions)

# Параметры задачи
n_inputs = 1
n_reservoir = 100
n_outputs = 1

# Создание сети
reservoir = ReservoirNetwork(n_inputs, n_reservoir, n_outputs)

# Генерация данных синуса
t = np.linspace(0, 50, 500)
data = np.sin(t)

# Подготовка данных для обучения
train_length = 400
test_length = 100

# Входные данные и цели для обучения
train_inputs = data[:train_length].reshape(-1, 1)
train_outputs = data[1:train_length + 1].reshape(-1, 1)

# Обучение
reservoir.train(train_inputs, train_outputs)

# Прогнозирование
test_inputs = data[train_length:train_length + test_length].reshape(-1, 1)
predictions = reservoir.predict(test_inputs)

# Визуализация результатов
plt.figure(figsize=(10, 5))
plt.plot(t[train_length:train_length + test_length], data[train_length:train_length + test_length], label="Истинное значение синуса")
plt.plot(t[train_length + 1:train_length + test_length + 1], predictions, label="Предсказание ESN")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.legend()
plt.show()
```
