```python
youtube.publishedAt = pd.to_datetime(youtube.publishedAt)
```

```python
youtube.publishedAt.dt.date
```

```python
import matplotlib.pyplot as plt
youtube_new = youtube.drop_duplicates(subset='video_id')
date = youtube_new['publishedAt'].dt.date.value_counts().sort_index()
x = date.index
plt.figure(figsize=(14, 5))
plt.title('-       ')
plt.xlabel(' ')
plt.ylabel('- ')
plt.grid(True)
plt.plot(x, date.values, marker='.', color='r', label=' ')
plt.xticks(x, fontsize=10, rotation=45)
plt.yticks(range(0,181, 10), fontsize=10)
plt.legend(fontsize=10)
plt.show()
```

### <font color='DarkOrange'>** 2 [0.5 a]**</font>

  `(bar plot)` -      .

```python
video_count = youtube['country'].value_counts().sort_index()
plt.figure(figsize=(14,5))
plt.title('-      ')
plt.xlabel('')
plt.ylabel('- ')
plt.grid(True)
plt.bar(video_count.index, video_count.values, color='pink')
plt.show()
```

### <font color='DarkOrange'>** 3 [1 ]**</font>

  `(scatter plot)`  -     -      (country='FR').          (comments_disabled=True).

    ?

```python
FR = youtube[(youtube['country'] == 'FR') & (youtube['comments_disabled'] == False)]
plt.figure(figsize=(14,5))
plt.title(' -   -     ')
plt.scatter(FR['view_count'],FR['comment_count'], marker='o', color = 'darkviolet')
plt.xlabel('- ')
plt.ylabel('- ')
plt.xscale('log')
plt.yscale('log')
plt.grid(True)
plt.show()
```
