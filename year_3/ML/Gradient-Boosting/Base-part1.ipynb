Значение MAE вычисляется как

$$
  MAE = \sum_{i = 1}^N\dfrac{|a(x_i) - y_i|}{N},
$$

где $N$ - число объектов в тестовой выборке, $x_i$ - вектор признаков i-го объекта, $a(x_i)$ - предсказание на i-ом объекте, $y_i$ - значение целевого признака на i-м объекте.

````python
datasets = {'A' : None, 'B': None, 'C': None}
# 1.
print('1. Количество  пропусков Nan/Null:', len(df[df.isnull().any(axis=1)]))
print('   Количество пропусков в  названии:', len(df.loc[(df.name == '') | (df.name == ' ')].name))
print()

# 2.
df['brand'] = df['name'].str.split().str[0]
print("2. После добавления столбца brand: ")
print(df.head())
print()

# 3.
category_names = ['Drive Type', 'Fuel Type', 'Transmission', 'name', 'Country', 'currency', 'brand']
print("3. Категориальные признаки: ", category_names)
for col in category_names:
  df[col] = df[col].astype('category')
print("   После конвертации")
df.info()
print()

# 4.
category_index = [df.columns.get_loc(col) for col in category_names]
datasets['A'] = {}
datasets['A']['X'] = df.drop(['price'], axis=1)
datasets['A']['y'] = df['price']

# 5.
datasets['B'] = {}
datasets['B']['X'] = df.drop(category_names, axis=1).drop(['price'], axis=1)
datasets['B']['y'] = df['price']

# 6.
datasets['C'] = {}
datasets['C']['X'] = pd.get_dummies(data=df.drop(['price'], axis=1), columns=category_names, dtype=int)
datasets['C']['y'] = df['price']

# 7.
for data in ['A', 'B', 'C']:
  X_train, X_test, y_train, y_test = train_test_split(datasets[data]['X'], datasets[data]['y'], test_size=0.25, random_state=0)
  datasets[data] = {}
  datasets[data]['X_train'] = X_train
  datasets[data]['X_test'] = X_test
  datasets[data]['y_train'] = y_train
  datasets[data]['y_test'] = y_test
````

````python
df_skl = pd.DataFrame(columns=['Dataset', 'Training time', 'Prediction time', 'Train MAE', 'Test MAE'])

for data in ['B', 'C']:
  model = GradientBoostingRegressor(**test_parameters)
  t = time.time()
  model.fit(datasets[data]['X_train'], datasets[data]['y_train'])
  time_fit = time.time() - t
  t = time.time()
  pred_train = model.predict(datasets[data]['X_train'])
  pred_test = model.predict(datasets[data]['X_test'])
  time_predict = time.time() - t
  mae_train = mean_absolute_error(datasets[data]['y_train'], pred_train)
  mae_test = mean_absolute_error(datasets[data]['y_test'], pred_test)
  result = [data, time_fit, time_predict, mae_train, mae_test]
  df_skl.loc[len(df_skl)] = result

df_skl
````

## Градиентный бустинг с XGBoost

<font color='CornflowerBlue'>**XGBoost**</font> стала достаточно популярной библиотекой, которая позволяет добиться хороших результатов без особых усилий (во всяком случае, так гласят легенды). И у нас для вас две новости: хорошая и плохая. Хорошая – xgboost, хоть и является сторонней библиотекой, сохраняет интерфейс sklearn. Плохая – она тоже не умеет из коробки работать с категориальными признаками. Ну что же, давайте пощупаем это безобразие. На этот раз нас интересует класс xgboost.XGBRegressor.

<font color='MediumTurquoise'>**Daily Fact:**</font> на самом деле в xgboost есть экспериментальная поддержка категориальных переменных, которая еще не добавлена в стабильную версию, поэтому ею мы пользоваться не будем.




### <font color='DarkOrange'>**Задание 3 [1.5 баллa]**</font>

2. Посчитать `MAE` для полученных предсказаний на соответствующих тренировочном и тестовом множествах (можно воспользоваться `sklearn.metrics.mean_absolute_error`). Замерить время предсказания
