````python
%matplotlib inline
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.metrics import mean_absolute_error, make_scorer

from hyperopt import hp, tpe, Trials
from hyperopt.fmin import fmin
from hyperopt.pyll import scope

from xgboost import XGBRegressor

from lightgbm import LGBMRegressor

from catboost import CatBoostRegressor

import matplotlib.pyplot as plt

import pandas as pd

import numpy as np

import time
````

````python
test_parameters = {"n_estimators": 1000, "max_depth": 5, "learning_rate":0.1}

df = pd.read_csv('/content/dataframe_YesIndex_YesHeader_C.csv', index_col=0)
df.head()
````

````python
datasets = {'A' : None, 'B': None, 'C': None}
# 1.
print('1.    Nan/Null:', len(df[df.isnull().any(axis=1)]))
print('       :', len(df.loc[(df.name == '') | (df.name == ' ')].name))
print()

# 2.
df['brand'] = df['name'].str.split().str[0]
print("2.    brand: ")
print(df.head())
print()

# 3.
category_names = ['Drive Type', 'Fuel Type', 'Transmission', 'name', 'Country', 'currency', 'brand']
print("3.  : ", category_names)
for col in category_names:
  df[col] = df[col].astype('category')
print("    ")
df.info()
print()

# 4.
category_index = [df.columns.get_loc(col) for col in category_names]
datasets['A'] = {}
datasets['A']['X'] = df.drop(['price'], axis=1)
datasets['A']['y'] = df['price']

# 5.
datasets['B'] = {}
datasets['B']['X'] = df.drop(category_names, axis=1).drop(['price'], axis=1)
datasets['B']['y'] = df['price']

# 6.
datasets['C'] = {}
datasets['C']['X'] = pd.get_dummies(data=df.drop(['price'], axis=1), columns=category_names, dtype=int)
datasets['C']['y'] = df['price']

# 7.
for data in ['A', 'B', 'C']:
  X_train, X_test, y_train, y_test = train_test_split(datasets[data]['X'], datasets[data]['y'], test_size=0.25, random_state=0)
  datasets[data] = {}
  datasets[data]['X_train'] = X_train
  datasets[data]['X_test'] = X_test
  datasets[data]['y_train'] = y_train
  datasets[data]['y_test'] = y_test
````

````python
df_skl = pd.DataFrame(columns=['Dataset', 'Training time', 'Prediction time', 'Train MAE', 'Test MAE'])

for data in ['B', 'C']:
  model = GradientBoostingRegressor(**test_parameters)
  t = time.time()
  model.fit(datasets[data]['X_train'], datasets[data]['y_train'])
  time_fit = time.time() - t
  t = time.time()
  pred_train = model.predict(datasets[data]['X_train'])
  pred_test = model.predict(datasets[data]['X_test'])
  time_predict = time.time() - t
  mae_train = mean_absolute_error(datasets[data]['y_train'], pred_train)
  mae_test = mean_absolute_error(datasets[data]['y_test'], pred_test)
  result = [data, time_fit, time_predict, mae_train, mae_test]
  df_skl.loc[len(df_skl)] = result

df_skl
````

````python
name = ['Sklearn', 'XGBoost', 'Lightgbm', 'CatBoost']
counter = 1
ids = np.arange(1, 5)
width = 0.5
plt.figure(figsize=(20, 16))
color = ''

for col in ['Training time', 'Prediction time', 'Train MAE', 'Test MAE']:
    if col == 'Training time':
        color = 'royalblue'
    elif col == 'Train MAE':
        color = 'tomato'
    elif col == 'Prediction time':
        color = 'forestgreen'
    else:
        color = 'violet'
    for data in ['A', 'B', 'C']:
        print_data = []
        for df_i in [df_skl, df_xgb, df_lightgbm, df_catboost]:
            if data in df_i['Dataset'].values:
                value = df_i[df_i['Dataset'] == data][col].iloc[0]
                print_data.append(value)
            else:
                print_data.append(0)
        plt.subplot(4, 3, counter)
        counter += 1
        plt.title(f'{col}  {data}', fontsize=16)
        plt.bar(ids - width / 2, print_data, width, color=color)
        plt.xlabel('', fontsize=14)
        plt.ylabel('', fontsize=14)
        plt.xticks(ids, name, rotation=20, fontsize=12)
        plt.grid()

plt.tight_layout()
plt.show()
````
